# Histograms / KDE


Histograms with KDE curves were used for **SalePrice**, **Lot Area**, and **Garage Area** to examine the distribution of these numeric features. These columns were selected because they are important continuous variables and can reveal patterns such as skewness, spread, and possible extreme values. This helps build an initial understanding of how the values are distributed.

```python
def show_distributions(df):
    for col in ["SalePrice", "Lot Area", "Garage Area"]:
        df[col].dropna().plot(kind="hist", bins=30, density=True, alpha=0.6)
        df[col].dropna().plot(kind="kde")
        plt.title(col)
        plt.show()
```

```python
show_distributions(df)
```

-------------------------------

# Grouped boxplots


Boxplots were created to compare **SalePrice** across categories in **House Style** and **Sale Condition**. These two columns were chosen because both may influence housing prices in different ways, and boxplots make those differences easy to compare visually. This visualization highlights variation, median differences, and possible outliers within each category.

```python
def plot_boxplots(df):
    plt.figure(figsize=(6,4))
    df.boxplot(column="SalePrice", by="House Style")
    plt.title("SalePrice by House Style")
    plt.suptitle("")
    plt.xlabel("House Style")
    plt.ylabel("SalePrice")
    plt.show()

    plt.figure(figsize=(6,4))
    df.boxplot(column="SalePrice", by="Sale Condition")
    plt.title("SalePrice by Sale Condition")
    plt.suptitle("")
    plt.xlabel("Sale Condition")
    plt.ylabel("SalePrice")
    plt.show()
```

```python
plot_boxplots(df)
```

------------------------------------------

# Correlation heatmap


A correlation heatmap was used to display the top numeric features most strongly related to **SalePrice**. The analysis focused on the ten most correlated variables so the plot remains clear and centered on the target variable. This makes it easier to identify which features have the strongest linear relationship with house prices.

```python
def correlation_heatmap(df):

    top_corr = df.corr(numeric_only=True)["SalePrice"].abs().drop("SalePrice").sort_values(ascending=False).head(10)

    top_features = top_corr.index.tolist() + ["SalePrice"]

    plt.figure(figsize=(8, 6))
    sns.heatmap(df[top_features].corr(), annot=True, cmap="coolwarm")
    plt.title("Top 10 Features Correlated with SalePrice")
    plt.show()
```

```python
correlation_heatmap(df)
```

-------------------------------------------

# Scatter plot


A scatter plot was used to study the relationship between **Gr Liv Area** and **SalePrice**, while **Overall Qual** was added as a color dimension. These columns were chosen because living area and overall quality are among the most influential housing features. This plot helps show not only the relationship between size and price, but also how quality affects that pattern.

```python
def scatter_relationship(df):
    plt.figure(figsize = (7, 5))
    scatter = plt.scatter(
        df["Gr Liv Area"],
        df["SalePrice"],
        c = df["Overall Qual"],
        alpha = 0.6
    )

    plt.title("Gr Liv Area vs SalePrice by Overall Quality")
    plt.xlabel("Gr Liv Area")
    plt.ylabel("SalePrice")
    plt.colorbar(scatter, label = "Overall Quality")
    plt.show()
```

```python
scatter_relationship(df)
```

----------------------------

# Groupby summary


A groupby summary was created for **Sale Condition** to calculate the average **SalePrice** in each category. This step was useful for comparing categories numerically instead of relying only on visual plots. It also identified which sale condition had the highest and lowest average price.

```python
def groupby_summary(df):
    summary = df.groupby("Sale Condition")["SalePrice"].mean().sort_values()

    print(summary)

    print("\nHighest group:")
    print(summary.idxmax(), summary.max())

    print("\nLowest group:")
    print(summary.idxmin(), summary.min())

    return summary
```

```python
groupby_summary(df)
```

--------------------------------

# run_eda


The run_eda() function was created to combine all analysis steps into one organized workflow. It runs the distributions, boxplots, heatmap, scatter plot, and grouped summary in sequence. This makes the notebook cleaner, easier to reuse, and more practical for repeating the full analysis.

```python
def run_eda(df):

    show_distributions(df)

    plot_boxplots(df)

    correlation_heatmap(df)

    scatter_relationship(df)

    groupby_summary(df)
```

```python
run_eda(df)
```
